## **2.1. Análisis para el potencial infinito**

### **2.1.1. Caja con una sola partícula**

Como bien sabemos, la ecuación de Schrödinger para este tipo de potenciales viene dada por:

$$
\frac{d^2 \Psi(X)}{dX^2} + k^2 \Psi(X) = 0 \tag{1}
$$


y, a su vez, esta ecuación permite una familia de soluciones del tipo:

$$
\Psi(X) = A' e^{ikX} + B' e^{-ikX} \Rightarrow \Psi(X) = A \sin(kX) + B \cos(kX). \tag{2}
$$

Cuando aplicamos las condiciones de frontera en \( X = 0 \) y \( X = L \), obtenemos:
$$
 \Psi(L) = A \sin(kL) = 0 \Rightarrow k_n L = n \pi \quad (n = 1, 2, 3, \dots).
$$

Y esta condición determina la dependencia entre la energía y el nivel:
$$
E_n = \frac{\hbar^2k_n^2}{2m} = \frac{n^2\hbar^2 \pi^2}{2mL^2} = \frac{h^2n^2}{8mL^2} \quad  (n = 1, 2, 3, \dots).
$$

Las condiciones de frontera anulan el coeficiente B en (2), y por ende, tenemos que la n-ésima solución de estado estacionario final para este tipo de sistemas, una vez normalizados los estados, tiene la forma:

$$
\Psi_n(X, t) = \phi_n(X) e^{-i E_n t / \hbar} \quad\text{donde, }\quad \phi_n(X) = \sqrt{\frac{2}{L}} \sin\left(\frac{n \pi X}{L}\right)
$$

Ahora, recordemos que la densidad de probabilidad para un estado estacionario es:

$$
P(X, t) = |\Psi(X, t)|^2 \quad\text{O sea, }\quad P_n(X) = |\phi_n(X)|^2 = \frac{2}{L} \sin^2\left(\frac{n \pi X}{L}\right) \tag{3}
$$

Hasta este punto se tiene la densidad de probabilidad para el sistema aislado. De alguna manera se debe modificar esta función para que incluya los efectos termodinámicos. Es aquí donde los autores del artículo introducen el factor de Boltzmann ($e^{-E_n / k_B T}$) a la densidad de probabilidad, y luego se realiza la sumatoria de todos los estados normalizados, llegando así al promedio térmico de la densidad de probabilidad, o probabilidad termalizada.

$$
P_{\text{th}}(X, T) = \frac{\sum_{n=1}^\infty P_n(x) e^{-E_n / k_B T}}{\sum_{n=1}^\infty e^{-E_n / k_B T}} \tag{4}
$$

nótese que esta función depende explícitamente de la posición y la temperatura. Reemplazando la ecuación (3) en (4), vemos que la probabilidad termalizada para una partícula es:

$$
P_{\text{th}}(X, T) = \frac{2}{L}  \frac{\sum_{n=1}^\infty \sin^2\left(\frac{n \pi X}{L}\right) e^{-E_n / k_B T}}{\sum_{n=1}^\infty e^{-E_n / k_B T}}
$$

Aunque matemáticamente no es muy compleja de interpretar o manipular, la expresión anterior es bastante engorrosa a la hora de escribir y de realizar cálculos analíticos y numéricos. Además, creería yo, también por razones de conveniencia computacional, los autores introducen dos cambios de variable, para llegar a una probabilidad termalizada que depende de coordenadas adimensionales. Todos los resultados computacionales que se reproducen en este mini-proyecto se realizan teniendo en cuenta este enfoque adimensional. Los cambios de variable pertinentes son los siguientes:

$$
t = \frac{8 m L^2k_B}{h^2}T, \quad x = \frac{\pi}{L}X, \quad \phi_n(x) = \sqrt{\frac{2}{\pi}} \sin(n x) \tag{I}
$$

El cambio de variable respecto a la temperatura, también se puede escribir como $t = \frac{k_B}{E_1}T $, así, tanto abajo como arriba hay unidades de energía, por lo tanto $t$ es adimensional. Por el lado de $x$, vemos que se tiene en cuenta la longitud de la caja y el factor $\pi$ que está dentro del argumento del seno, simplificando a su vez los autoestados base de la función de onda.

Para empezar a reproducir los resultados computacionales del artículo, teniendo en cuenta que el cambio de variable redefine el intervalo espacial, se debe definir el dominio de las f.d.p. así:

```python
x = np.linspace(0, np.pi, 100)
```
 

A su vez, se deben definir las autofunciones que hacen parte de la base de autoestados de la función de onda:



```python
def phi(n, x):
    c = np.sqrt(2 / np.pi)
    return c*np.sin(n*x)
```



Una vez realizadas las sustituciones, nuestra nueva "función computacional" adimensional que representa la probabilidad termalizada está dada por:

$$
P_{\text{th}}(x, t) = \frac{2}{\pi}\frac{\sum_{n=1}^\infty \sin^2(n x) e^{-n^2 / t}}{\sum_{n=1}^\infty e^{-n^2 / t}} \tag{5}
$$

Cuyo código sería algo más o menos así:

```python
def P_th(x, t, nmax):
    suma1 = 0
    suma2 = 0
    for i in range(1, nmax+1):
        p = phi(i, x) * phi(i, x)
        term1 = BoltzmannFactor(i, t) * p
        suma1 += term1
        suma2 += BoltzmannFactor(i, t)
    Total = suma1 / suma2
    
    return Total

```

### **2.1.2. Caja con dos partículas sin spin**

Ya vimos el caso de una sola partícula dentro de la caja caliente, sin considerar el spin. Ahora, hagamos lo propio para el caso de dos partículas, también sin spin.

Recordando que los fermiones cumplen el principio de exclusión de Pauli, entonces se puede ver que la función de onda total para estas dos partículas debe ser antisimétrica ante el intercambio de partículas. Es decir, los niveles de energía no pueden ser iguales, no existe un estado físico que represente este caso (Por lo tanto la función se debe anular cuando $n_1 = n_2$). Con todo esto en cuenta, y considerando el caso sin spin, la función de onda total para dos fermiones es:


$$
\Psi^f_{n_1, n_2}(x_1, x_2) = \frac{1}{\sqrt{2}} \left[\phi_{n_1}(x_1) \phi_{n_2}(x_2) - \phi_{n_2}(x_1) \phi_{n_1}(x_2)\right] \tag{6}
$$

Por su parte, los bosones no presentan las mismas restricciones en los estados observadas en los fermiones, y la función de onda total para dos bosones indistinguibles debe ser simétrica ante el intercambio de partículas, teniendo entonces la siguiente representación de esta fucnión de onda:

$$
\Psi^b_{n_1, n_2}(x_1, x_2) = \frac{1}{\sqrt{2}} \left[\phi_{n_1}(x_1) \phi_{n_2}(x_2) + \phi_{n_2}(x_1) \phi_{n_1}(x_2)\right] \tag{7}
$$

Una forma de construir estas funciones de onda por tipo de partícula sería la siguiente:

```python
def psi(n, x, mode):
    c = np.sqrt(1/2)
    term1 = phi(n[0], x[0]) * phi(n[1], x[1])
    term2 = phi(n[0], x[1]) * phi(n[1], x[0])

    if mode == 'fermion':
        Total = c * (term1 - term2)
    elif mode == 'boson':
        Total = c * (term1 + term2)
         
    return Total
```

Una vez se tiene la función de onda para estas partículas fermiónicas y bosónicas, podemos llegar a una expresión para la densidad de probabilidad, usando argumentos similares a los del caso de una partícula. En este caso, tenemos que la función densidad probabilidad de encontrar un fermión en la posición $x_1$ y otro en la posición $x_2$ viene dada por:

$$
P^f_{n_1, n_2}(x_1, x_2) = \frac{2}{\pi^2} \left[\sin(n_1 x_1) \sin(n_2 x_2) - \sin(n_1 x_2) \sin(n_2 x_1)\right]^2 \tag{8}
$$

Por su parte, la función densidad probabilidad de encontrar un bosón en la posición $x_1$ y otro en la posición $x_2$ sería:

$$
P^b_{n_1, n_2}(x_1, x_2) = \frac{2}{\pi^2} \left[\sin(n_1 x_1) \sin(n_2 x_2) + \sin(n_1 x_2) \sin(n_2 x_1)\right]^2 \tag{9}
$$

Es importante resaltar, como lo hacen los autores del artículo, que las ecuaciones (6) a (9) están todas escritas en términos de las variables adimensionales, tomando mano de las expresiones vistas en la ecuación $(I)$.

La construcción computacional de esta función tiene una forma elegante:

```python
def P(n, x, mode):
    p = psi(n, x, mode) * psi(n, x, mode)
    return p
```

Ahora, estas ecuaciones representan la función densidad de probabilidad de dos partículas en una caja aislada, y no hemos tenido en cuenta las propiedades térmicas, cuando ponemos la caja en contacto con el reservorio de partículas caliente. Para ello, se debe realizar un procedimiento análogo al que se realiza para la probabilidad termalizada de una sola partícula. Debemos tener en cuenta la f.d.p. para las dos partículas, y añadir el factor de Boltzmann a la ecuación, posteriormente realizar la sumatoria sobre todos los posibles estados y, finalmente, normalizar, para así obtener el promedio térmico de la probabilidad. Estas consideraciones se reducen a la siguiente fórmula:

$$
P_{th}(x_1, x_2, t) =
\frac{
\sum_{n_1=1}^\infty \sum_{n_2=1}^\infty P_{n_1,n_2}(x_1, x_2) \exp\left(-\frac{E_{n_1,n_2}}{t}\right)
}{
\sum_{n_1=1}^\infty \sum_{n_2=1}^\infty \exp\left(-\frac{E_{n_1,n_2}}{t}\right)
} \quad \text{con: } \quad E_{n_1, n_2} = n_1^2 + n_2^2 \tag{II}
$$

Como se describe en el artículo, se debe tener sumo cuidado al evaluar la probabilidad termalizada. Primero, porque el cálculo computacional requiere que el límite superior de las sumas sea reemplazado con un parámetro nmax de valor finito a determinar. Por otro lado, se debe evitar agregar términos repetidos a la sumatoria. Además, como mencionan los autores, debemos recordar que dos fermiones no pueden ocupar el mismo nivel. Con estas consideraciones en mente, y teniendo en cuenta las fórmulas (8) y (9), la expresión $(II)$ asume la siguiente forma para fermiones o bosones:


**Fermiones:**

$$
P_{\text{th}}^f(x_1, x_2, t) = \frac{2}{\pi^2} \frac{\sum_{n_1=1}^{n_{\text{max}}-1} \sum_{n_2=n_1+1}^{n_{\text{max}}} \left[\sin(n_1 x_1) \sin(n_2 x_2) - \sin(n_1 x_2) \sin(n_2 x_1)\right]^2 e^{-(n_1^2 + n_2^2) / t}}{\sum_{n_1=1}^{n_{\text{max}}-1} \sum_{n_2=n_1+1}^{n_{\text{max}}} e^{-(n_1^2 + n_2^2) / t}} \tag{10}
$$

**Bosones:**

$$
P_{\text{th}}^b(x_1, x_2, t) = \frac{2}{\pi^2}\frac{\sum_{n_1=1}^{n_{\text{max}}} \sum_{n_2=n_1}^{n_{\text{max}}} \left[\sin(n_1 x_1) \sin(n_2 x_2) + \sin(n_1 x_2) \sin(n_2 x_1)\right]^2 e^{-(n_1^2 + n_2^2) / t}}{\sum_{n_1=1}^{n_{\text{max}}} \sum_{n_2=n_1}^{n_{\text{max}}} e^{-(n_1^2 + n_2^2) / t}} \tag{11}
$$

Se podría reducir todo este aparato matemático a una forma computacional también elegante y sencilla, que sería así:

```python
def PP_th(x, t, nmax, mode):
    suma1 = 0
    suma2 = 0
    for i in range(1, nmax):
        for j in range(i+1, nmax+1):
            term1 = BoltzmannFactor([i, j], t) * P([i, j], [x[0], x[1]], mode)
            suma1 += term1
            suma2 += BoltzmannFactor([i, j], t)
    
    Total = suma1 / suma2

    return Total
```

Aunque, como se verá más adelante (y en la construcción del código), esta función debe ser modificada según algunos requerimientos.

### **2.1.3. Fermiones con spin**

Llegamos a la última sección del artículo. En esta parte, los autores se proponen realizar el análisis tomando en cuenta la contribución del spin a la función de onda completa de los fermiones. Cuando se toman en cuenta estas contribuciones, la función de onda para fermiones se puede separar en dos componentes, una simétrica espacial y otra antisimétrica espacial (a diferencia del caso anterior, donde únicamente existe la parte antisimétrica para garantizar el principio de exclusión de Pauli), cada una acompañada por su contraparte referente al spin. Además, la contribución del spin a la parte simétrica se hace mediante un singlete, mientras que la contribución del spin a la parte antisimétrica se hace con un triplete. Con esto en cuenta, sólo hay una forma de expresar la función de onda simétrica para fermiones con spin, y puede ser escrita como sigue:

$$
\Psi^f_{n_1, n_2}(x_1, x_2) = \frac{1}{\sqrt{2}} \left[\phi_{n_1}(x_1) \phi_{n_2}(x_2) + \phi_{n_2}(x_1) \phi_{n_1}(x_2)\right] \left[\chi_+(1) \chi_-(2) - \chi_+(2) \chi_-(1)\right]
$$

Por otro lado, las tres formas de escribir su parte antisimétrica espacial son:

$$
\Psi^f_{n_1, n_2}(x_1, x_2) = \frac{1}{\sqrt{2}} \left[\phi_{n_1}(x_1) \phi_{n_2}(x_2) - \phi_{n_2}(x_1) \phi_{n_1}(x_2)\right] \left[\chi_+(1) \chi_-(2) + \chi_+(2) \chi_-(1)\right]
$$

$$
\Psi^f_{n_1, n_2}(x_1, x_2) = \frac{1}{\sqrt{2}} \left[\phi_{n_1}(x_1) \phi_{n_2}(x_2) - \phi_{n_2}(x_1) \phi_{n_1}(x_2)\right] \left[\chi_+(1) \chi_+(2)\right]
$$

$$
\Psi^f_{n_1, n_2}(x_1, x_2) = \frac{1}{\sqrt{2}} \left[\phi_{n_1}(x_1) \phi_{n_2}(x_2) - \phi_{n_2}(x_1) \phi_{n_1}(x_2)\right] \left[\chi_-(1) \chi_-(2)\right]
$$

Ahora, en el artículo se plantea una duda que sale directamente al ver esta forma de representar la función de onda completa de los dos fermiones con spin: ¿Cómo calculamos la probabilidad de encontrar una partícula en la posición $x_1$ y otra en $x_2$, sin que la forma funcional de esta dependa explícitamente del spin?

La respuesta es la siguiente: "será necesario tener en cuenta las cuatro funciones de onda posibles y recordar que tres de ellas están asociadas a la función de onda espacial antisimétrica y una a la función de onda espacial simétrica". Con esta respuesta en mente, la fórmula para la densidad de probabilidad de los dos fermiones con spin es:

$$
P^{f_s}_{n_1, n_2}(x_1, x_2) = \frac{1}{4} \left(\frac{1}{2}\left[ \phi_{n_1}(x_1) \phi_{n_2}(x_2) + \phi_{n_2}(x_1) \phi_{n_1}(x_2) \right]^2\right) + \frac{3}{4} \left(\frac{1}{2}\left[ \phi_{n_1}(x_1) \phi_{n_2}(x_2) - \phi_{n_2}(x_1) \phi_{n_1}(x_2) \right]^2\right) \tag{12}
$$

El último paso que se debería dar, es el de convertir el sistema aislado en uno con propiedades térmicas que dependen del reservorio de partículas. Entonces, como en los casos anteriores, cuando ponemos la caja con dos fermiones con spin en contacto con el reservorio caliente, su probabilidad termalizada queda en la forma:


$$
P^{f_s}_{th}(x_1, x_2, t) =
\frac{
\sum_{n_1=1}^{n_{max} - 1} \sum_{n_2=n_1 + 1}^{n_{max}} P^{f_s}_{n_1,n_2}(x_1, x_2) \exp\left(-\frac{n_1^2 + n_2^2}{t}\right)
}{
\sum_{n_1=1}^{n_{max} - 1} \sum_{n_2=n_1 + 1}^{n_{max}} \exp\left(-\frac{n_1^2 + n_2^2}{t}\right)
} \tag{13}
$$

Para reproducir los resultados computacionales en este caso, la probabilidad termalizada tiene una forma similar a la que se exploró en la sección anterior. Sin embargo, la f.d.p. de los dos fermiones dentro de la caja cambia considerablemente a la hora de considerar el spin. La forma de la función computacional para estos fermiones podría ser así:

```python
def P_s(n, x):
    term1 = psi(n, x, 'boson') * psi(n, x, 'boson') / 2
    term2 = psi(n, x, 'fermion') * psi(n, x, 'fermion') / 2
    p = (term1 / 4) + (3*term2 / 4)
    
    return p
```

## **2.2. Análisis para el potencial finito**

Así como el pozo de potencial infinito, el potencial finito es también un problema bien estudiado y resuelto. En estos casos de caja de tamaño finito, tenemos dos posibles análisis teniendo en cuenta la condición $E>V_0$ (estados de scattering) o $E<V_0$ (estados ligados). En ambos casos, el potencial involucrado en el problema se puede escribir de la siguiente manera:

$$
V(X) = 
\begin{cases} 
V_0, & ~~~ X < -L/2 \\
0, & ~~~ -L/2 \leq X \leq L/2, \\
V_0, & ~~~ X > L/2,
\end{cases}
$$

Comencemos analizando primero la condición de estados ligados. Para el caso dentro de la caja, el potencial se anula y la ecuación de Schrödinger a resolver es la misma que se tiene en $(1)$, con $k = \sqrt{\frac{2mE}{\hbar^2}}$. Para el caso fuera de la caja, la ecuación $(1)$ se tranforma en:

$$
\frac{d^2 \Psi(X)}{dX^2} - \kappa^2 \Psi(X) = 0, \text{ con} \quad \kappa = \sqrt{\frac{2m(V_0 - E)}{\hbar^2}} \tag{14}
$$

A partir de estas ecuaciones, tenemos soluciones para las tres regiones definidas por el potencial:

$$
\Psi_I(X) = Ae^{\kappa X}, \quad \Psi_{III}(X) = De^{-\kappa X} \tag{15}
$$

$$
\Psi_{II}(X) = B\cos{(kX)} + C\sin{(kX)} \tag{16}
$$

Ahora, como el potencial finito es simétrico y el operador paridad conmuta con el hamiltoniano, debemos reflejar estos hechos (acompañados de las condiciones de continuidad) en las soluciones para que estas tengan sentido físico. Entonces, las soluciones se deben separar en caso simétrico (o par) y asimétrico (impar), obteniendo entonces:

$$
\Psi_s(X) =
\begin{cases} 
Ae^{\kappa X}, & X < -L/2, \\
B \cos(kX), & -L/2 \leq X \leq L/2, \\
Ae^{-\kappa X}, & X > L/2.
\end{cases} \tag{17}
$$ 


$$
\Psi_a(X) =
\begin{cases} 
Ae^{\kappa X}, & X < -L/2, \\
C \sin(k X), & -L/2 \leq X \leq L/2, \\
-Ae^{-\kappa X}, & X > L/2,
\end{cases} \tag{18}
$$

Las constantes de normalización deben tener esta forma también por razones de paridad. En los extremos de la caja se debe cumplir la continuidad de la función de onda completa, es decir que $\Psi_I(-L/2) = \Psi_{II}(-L/2)$, y $\Psi_{II}(L/2) = \Psi_{III}(L/2)$. 

Por ser una ED de segundo orden, las primeras derivadas también deben cumplir las condiciones de continuidad, por tanto $\Psi'_I(-L/2) = \Psi'_{II}(-L/2)$, y $\Psi'_{II}(L/2) = \Psi'_{III}(L/2)$. Estas condiciones se pueden escribir mejor en términos de la derivada logarítmica, imponiendo así condiciones de continuidad en las derivadas logarítmicas de ambas regiones:

$$
\frac{\Psi'_I(-L/2)}{\Psi_I(-L/2)} = \frac{\Psi'_{II}(-L/2)}{\Psi_{II}(-L/2)} \quad\text{y, }\quad \frac{\Psi'_{II}(L/2)}{\Psi_{II}(L/2)} = \frac{\Psi'_{III}(L/2)}{\Psi_{III}(L/2)} \tag{19}
$$

Entonces, para soluciones simétricas, estas ecuaciones quedan de la siguiente manera, (Al aplicar ambas condiciones de continuidad, se llega a la misma expresión):

$$
\tan\left(\frac{kL}{2}\right) = \frac{\kappa}{k} \tag{a}
$$

Por su parte, al aplicar las condiciones de las ecuaciones $(19)$ en las soluciones asimétricas llevan a la siguiente ecuación trascendental, también para ambas condiciones:

$$
\cot\left(\frac{kL}{2}\right) = -\frac{\kappa}{k} \tag{b}
$$

Ahora, para poder resolver estas ecuaciones trascendentales, debemos realizar un desarrollo previo que facilite su tratamiento, hagámoslo para la ecuación $(a)$ y un procedimiento análogo servirá para la ecuación $(b)$. Multipliquemos y dividamos por $L/2$, y luego expandamos el valor de $\kappa$ para obtener:

$$\tan\left(\frac{kL}{2}\right) = \frac{\sqrt{\frac{2mV_0(L/2)^2}{\hbar^2} - \frac{2mE(L/2)^2}{\hbar^2}}}{k(L/2)} \quad\implies k(L/2)\tan\left(\frac{kL}{2}\right) =\sqrt{\frac{2mV_0(L/2)^2}{\hbar^2} - k^2(L/2)^2} $$

Y si tomamos las sustituciones $R^2 = \frac{2mV_0(L/2)^2}{\hbar^2}$, y $\alpha_n = \frac{k_nL}{2}$, entonces podemos reescribir las ecuaciones trascendentales como:

$$
\alpha_n\tan\left(\alpha_n\right) = \sqrt{R^2 - \alpha_n^2} \quad\text{ y, }\quad -\alpha_n\cot\left(\alpha_n\right) = \sqrt{R^2 - \alpha_n^2} \tag{20}
$$

Tomando en cuenta los cambios de variables realizados, las soluciones dentro de la caja serían de la forma:

$$
B \cos(\frac{2\alpha_n}{L}X) \quad\text{ Estados pares, } \quad C \sin(\frac{2\alpha_n}{L}X) \quad\text{ Estados impares }
$$

Ahora, si queremos buscar los valores de las constantes de normalización de, por ejemplo los estados simétricos, debemos hallar la norma al cuadrado de la función de onda en las tres regiones, teniendo que:

$$
\int_{-\infty}^{\infty}|\Psi(X)|^2dX = \int_{-\infty}^{-L/2}|\Psi_I(X)|^2dX + \int_{-L/2}^{L/2}|\Psi_{II}(X)|^2dX + \int_{L/2}^{\infty}|\Psi_{III}(X)|^2dX = 1
$$

$$
1 = \int_{-\infty}^{-L/2}|A|^2e^{2\kappa X}dX + \int_{-L/2}^{L/2}|B|^2\cos^2\bigl(\frac{2\alpha_nX}{L}\bigr)dX + \int_{L/2}^{\infty}|A|^2e^{-2\kappa X}dX
$$





Esta condición nos lleva a la siguiente relación entre las constantes de normalización dentro y fuera de la caja de potencial finito:

$$
|B|^2 = \frac{2(2\alpha_n)}{\bigl[\sin{(2\alpha_n) + 2\alpha_n}\bigr]L}\bigl[1 - \frac{|A|^2}{\kappa}e^{-\kappa L}\bigr]
$$

Ahora, para efectos computacionales, realicemos el cambio de variable: $x = \frac{2}{L}X$, entonces tenemos las siguientes soluciones pares:

$$
\Psi_s(x) =
\begin{cases} 
Ae^{\frac{L\kappa_n}{2}x }, & x < -1, \\
B \cos(\alpha_n x), & -1 \leq x \leq 1, \\
Ae^{-\frac{L\kappa_n}{2}x }, & x > 1.
\end{cases} \tag{21}
$$

Tenemos todo listo para definir la densidad de probabilidad de este sistema:

$$
P^s_n(x) =
\begin{cases} 
|A|^2e^{L\kappa_n x }, & x < -1, \\
|B|^2 \cos^2(\alpha_n x), & -1 \leq x \leq 1, \\
|A|^2e^{-L\kappa_n x }, & x > 1.
\end{cases} \tag{22}
$$


---

### **2.2.1. Caja finita con una sola partícula**

Con la ecuación $(22)$ tenemos la densidad de probabilidad en la caja de potencial finito, a partir de ella podemos empezar a construir la probabilidad termalizada de la caja finita. La idea es aplicar desarrollos similares a los expuestos en secciones anteriores. Teniendo entonces que 

$$
P^s_{\text{th}}(x, T) = \frac{\sum_{n=1}^\infty P^s_n(x) e^{-E_n / k_B T}}{\sum_{n=1}^\infty e^{-E_n / k_B T}} \tag{4}
$$

Por la naturaleza de este problema, el análisis de la probabilidad termalizada es un poco más complicado, pues se deben tener algunas consideraciones en cuenta.

````python
def f_s(alpha_n, R):
  return alpha_n * np.tan(alpha_n) - np.sqrt(R**2 - alpha_n**2)
                    
def f_a(alpha_n, R):
  return (alpha_n / np.tan(alpha_n)) + np.sqrt(R**2 - alpha_n**2)

R_a = 3*np.pi/2

roots_s = finiteBox(0).find_roots(f_s, R_a, symmetry=True)  # Raíces para modo simétrico
roots_a = finiteBox(0).find_roots(f_a, R_a, symmetry=False)  # Raíces para modo asimétrico
````

````python
# Rango de x
x = np.linspace(-1, 1, 500)

# Parámetros
A = 1  # Amplitud arbitraria
u = 0.5  # Parámetro de decaimiento (ejemplo)

# Evaluar phi(x) para modos simétrico y asimétrico
phi_s = finiteBox(0).phi(A, roots_s[0], u, x, symmetry=True)
phi_a = finiteBox(0).phi(A, roots_a[0], u, x, symmetry=False)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x, phi_s, label="Modo Simétrico")
plt.plot(x, phi_a, label="Modo Asimétrico")
plt.axvline(x=-1, color='gray', linestyle='--', label="Bordes del pozo")
plt.axvline(x=1, color='gray', linestyle='--')
plt.xlabel("x")
plt.ylabel("φ(x)")
plt.title("Funciones de onda para el pozo de potencial finito")
plt.legend()
plt.grid(True)
plt.show()
````
